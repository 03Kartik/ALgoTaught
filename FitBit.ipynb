```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('FitBit data.csv')
df.shape
```

```python
df.isnull().count()
```

```python
df['ActivityDate']=pd.to_datetime(df['ActivityDate'])
df['week']=df['ActivityDate'].dt.dayofweek
df['month']=df['ActivityDate'].dt.month
df['year']=df['ActivityDate'].dt.year
df.head()
```

```python
np.unique(df['month'])
```

```python
(df['TotalDistance'] == df['TrackerDistance']).count()
```

```python
drop=['ActivityDate','TotalDistance','year']
df.drop(drop,inplace=True,axis=1)
df.head()
```

```python
df.describe()
```

```python
df['month'].value_counts().plot.bar()
```

In April There Is A Lot Activity Than March

```python
df.groupby('month')['Id'].count()
```

```python
df.set_index('Id').plot(subplots=True, figsize=(10,50))
```

```python
df.groupby('Id')['TrackerDistance'].sum().sort_values().plot.bar(color = 'Brown', figsize = (20, 7))
plt.title(' Total Distance Covered By Various Id')
plt.xlabel('Id')
plt.ylabel('Total Distance Travelled')
plt.show
```

## Id 887768931 Travelled Most


```python
df.groupby('Id')['TotalSteps'].sum().sort_values().plot.bar(color = 'Blue', figsize = (20, 7))
plt.title('Total Steps Walked By Various Id')
plt.xlabel('Id')
plt.ylabel('Total Steps Travelled')
plt.show
```

## Id 1503960366 Walk Most Step In The Entire Dataset

```python
df.groupby('Id')['Calories'].sum().sort_values().plot.bar(color = 'Grey', figsize = (20, 7))
plt.title(' Total Calories Burn')
plt.xlabel('Id')
plt.ylabel('Total Calories Burned')
plt.show
```

### Id 4020332650 Burns More Calories Than Others
