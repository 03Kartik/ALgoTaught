```python
tim = ['day','month']
for i in tim:
    print(pd.DataFrame(df.groupby(i)['area'].sum()).plot.bar())
```

## It Is Clearly Shown In The Graph Than Forest Get Maximum Area Damage In The Month Of September And August As Per Comparision From Other Monnths And On Specially On Saturday

```python
df.columns
```

```python
per = ['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH','wind', 'rain']
for i in per:
    print(pd.DataFrame(df[df['month'] =='sep'].groupby(i)['area'].sum()).plot.bar(figsize=(16,9)))
```

```python
per = ['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH','wind', 'rain']
for i in per:
    plt.plot(df[df['month'] =='aug'].groupby(i)['area'].sum(),label=i +' In August')
    plt.plot(df[df['month'] =='sep'].groupby(i)['area'].sum(),label=i + ' In September')
    plt.plot(df[df['month'] =='jan'].groupby(i)['area'].sum(),label=i + ' In January')
    plt.plot(df[df['month'] =='jun'].groupby(i)['area'].sum(),label=i + ' In June')

    plt.ylabel('Damaged Area')
    plt.xlabel('Value Of '+i)
    plt.legend()
    plt.show()
```

## Here We FInd DIfferrent Parameter Which Cause Heavy Damage In The Month Of Sep And August

```python
per = ['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH','wind', 'rain']
for i in per:
    plt.plot(df[df['day'] =='sat'].groupby(i)['area'].sum(),label=i +' In August')
    plt.plot(df[df['day'] =='fri'].groupby(i)['area'].sum(),label=i +' In August')

    plt.ylabel('Damaged Area')
    plt.xlabel('Value Of '+i)
    plt.legend()
    plt.show()
```
